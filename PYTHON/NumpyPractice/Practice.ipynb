multi_dimention Array


```python
dimension=np.array((['Jan','Feb','Mar'],
                   [300,400,500],
                   [600,700,800],
                   [50,30,200]))
Sum=np.sum(dimension)
print(Sum)
print(dimension.shape)
print(dimension)
print(dimension[2][2])
print(dimension[1,:])
print(dimension[:,1])
```

Highest rate customers

```python
rating=np.array([2,3,4.5,2,6])
C_id=np.array(['c101','c102','c103','c104'])
print(np.where(C_id=='c103'))
print(rating[np.where(C_id=='c103')])
print(np.sort(rating)[::-1])
print(np.max(rating))
print(C_id[np.where(rating==4.5)])


```
